```python
import pandas as pd
import networkx as nx
import random
```

```python
def create_graph(lists):
    df = pd.DataFrame(lists, columns=['source', 'target'])
    G = nx.from_pandas_edgelist(df, 'source', 'target')
    return G
def create_dataframe(num_rows, upper_limit):
    df = pd.DataFrame({'col1': [random.randint(1, upper_limit) for _ in range(num_rows)],
                       'col2': [random.randint(1, upper_limit) for _ in range(num_rows)]})
    tuples = [tuple(x) for x in df.to_numpy()]
    return tuples
```

```python
tuples = create_dataframe(100,30)
G = create_graph(tuples)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=60, font_size=10, font_weight='bold')
```
